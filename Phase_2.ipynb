```
# Dataset metadata processing
max_length = 0
docs = []
index = 0

with open(text_path, encoding = 'utf-8') as fp:
  for line in fp:
    parts = line.strip().split('|')
    max_length = max(max_length, len(parts[2]))
    docs.append(parts[2])
    index += 1

print('maximum length of charcters in a sentence of training data is', max_length)
print('total length ', index)
print(docs[:5])
```

```
# Model Development
# part-1

# preparing tokenizer
t = Tokenizer()
t.fit_on_texts(docs) 
vocab_size = len(t.word_index) + 1
print('vocab size ', vocab_size)

encoded_docs = t.texts_to_sequences(docs)
print('encoded docs: ', encoded_docs)

# padding the documents to a max words containing sentence
padded_docs = pad_sequences(encoded_docs, maxlen = max_length, padding = 'post')
# padded_docs = docs
print('padded docs: ', padded_docs)

# loading the whole embedding into memory
embeddings_index = dict()
f = open('/content/gdrive/My Drive/Dataset/Test/Glove/glove.6B.100d.txt')
for line in f:
	values = line.split()
	word = values[0]
	coefs = asarray(values[1:], dtype = 'float32')
	embeddings_index[word] = coefs
f.close()
print('Loaded %s word vectors.' % len(embeddings_index))

# create a weight matrix for words in training docs
embedding_matrix = zeros((vocab_size, 100))
for word, i in t.word_index.items():
  embedding_vector = embeddings_index.get(word)
  if embedding_vector is not None:
    embedding_matrix[i] = embedding_vector
```

```
output = np.load('/content/gdrive/My Drive/Dataset/Test/Mel-Numpy/mel.npy')[:2016]
print(output.shape)
inp = padded_docs[:2016]
print(inp.shape)
print(padded_docs[:2043])
```

```
# pipeline-1 -> Text Encoding
# model-1

model1 = Sequential()
model1.add(Embedding(vocab_size, 100,  weights=[embedding_matrix], input_length = 187, trainable=False))
# # model1.add(Flatten())
# layer_output = [layer.output for layer in model1.layers]
# print(model1.summary())
model1.add(Convolution1D(filters=256, kernel_size=(5), activation='relu'))
model1.add(Convolution1D(filters=128, kernel_size=(5), activation='relu'))
model1.add(Convolution1D(filters=64, kernel_size=(3), activation='relu'))
# print(layer_output)
# print(layer_output[0].get_shape().as_list())
model1.add(Bidirectional(LSTM(200)))
model1.add(Reshape((40,10)))
# model1.add(UpSampling2D())
# model1.add(UpSampling2D())

# opt = keras.optimizers.Adam(lr = 0.0001)
model1.compile(loss='mse',optimizer='adam',metrics=['acc'])
# summarize the model1
print(model1.summary())

# model1.fit(inp, output, batch_size=32, epochs=30)
```

```
# pipeline-2 -> user/speaker encoding
# model-2

model2 = Sequential()
# model2.add(keras.Input(shape=(128,128,3)))
# model2.add(Flatten())
# model2.add(keras.layers.Dense(128,activation='relu'))
# model2.add(keras.layers.Dense(256, activation='relu'))
# model2.add(keras.layers.Dense(128, activation='relu'))
model2.add(Convolution2D(32, 5 ,data_format = 'channels_last', padding = 'same', activation = 'relu', strides = 1, input_shape = (80, None, 1)))  
model2.add(AveragePooling2D(pool_size = (2, 2), padding = 'same'))  

model2.add(Convolution2D(64, 3, padding = 'same'))  
model2.add(AveragePooling2D(pool_size = (2, 2), padding = 'same'))  

layer_output = [layer.output for layer in model2.layers]
print(layer_output)
# print(layer_output[3].get_shape().as_list()[1:3])
# print(layer_output[0].get_shape().as_list()[1:3])
model2.add(Conv2DTranspose(24,(layer_output[3].get_shape().as_list()[1:3]),padding = 'same', activation = 'relu', strides = 2))
model2.add(Conv2DTranspose(3,(layer_output[0].get_shape().as_list()[1:3]),padding = 'same', activation = 'relu', strides = 2))

model2.compile(optimizer='adam', loss='mae', metrics=['accuracy'])
print(model2.summary())

model2.fit(orig, orig, epochs = 30)
```

```
# merging two pipelines of models and preperation of final model

from keras.layers import *
from keras.models import Model

merged = Multiply()([model1.output, model2.output])
# print(mergedOut.shape)
# mergedNew = Conv2D(32, 5, padding = 'same', activation = 'relu', strides = 1)(mergedOut)
# print(mergedOut.shape)

# finalModel = Sequential()
# finalModel.add(combinedModel)
finalModel = Convolution2D(24, 5, padding = 'same', activation = 'relu', strides = 1)(merged)
finalModel = AveragePooling2D(pool_size = (2, 2), padding = 'same')(finalModel)

finalModel = Convolution2D(48, 3, padding = 'same')(finalModel)
finalModel = AveragePooling2D(pool_size = (2, 2), padding = 'same')(finalModel) 

finalModel = Convolution2D(64, 3, padding = 'same')(finalModel)
finalModel = AveragePooling2D(pool_size = (2, 2), padding = 'same')(finalModel) 
# layer_output = [layer.output for layer in finalModel.layers]
# print(layer_output)
# print(layer_output[3].get_shape().as_list()[1:3])
# print(layer_output[0].get_shape().as_list()[1:3])

finalModel = Conv2DTranspose(48,(finalModel.shape[1:3]),padding = 'same', activation = 'relu', strides = 2)(finalModel)
finalModel = Conv2DTranspose(24,(finalModel.shape[1:3]),padding = 'same', activation = 'relu', strides = 2)(finalModel)
finalModel = Conv2DTranspose(3,(finalModel.shape[1:3]),padding = 'same', activation = 'relu', strides = 2)(finalModel)

combinedModel = Model([model1.input,model2.input], finalModel)
combinedModel.compile(optimizer='adam', loss='mae', metrics=['accuracy'])

print(combinedModel.summary())

combinedModel.fit([inp, output], output, epochs = 30)
```
